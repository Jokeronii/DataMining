```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/tugas datmin/Data Covid19.csv")
dataset
```

```python
dataset.info()
```

```python
dataset['Last Update'].unique()
```

```python
last_update='28/02/2021 05.22'
dataset[dataset['Last Update']==last_update]
```

```python
unique_countries = dataset[dataset.Confirmed>0]['Country/Region'].unique()
unique_countries.sort()
unique_countries
```

```python
country_confirmed_case = []
for i in unique_countries:
  country_confirmed_case.append(dataset[dataset.Confirmed>0][dataset['Country/Region']==i][dataset['Last Update']==last_update].Confirmed.sum())
```

```python
print(set(zip(unique_countries,country_confirmed_case)))
```

```python
unique_province = dataset['Province/State'][dataset.Confirmed>1].unique()
unique_province
```

```python
province_confirmed_case = []
for i in unique_province:
  province_confirmed_case.append(dataset[dataset.Confirmed > 0][dataset['Province/State']==i][dataset['Last Update']==last_update].Confirmed.sum())
```

```python
print(set(zip(unique_province,province_confirmed_case)))
```

cek ada data yg nan atau tidak,jika ada ditampung pada array nan_indices

```python
nan_indices = []
for i in range(len(unique_province)):
  if type(unique_province[i])==float:
    nan_indices.append(i)
```

```python
unique_province = list(unique_province)
province_confirmed_case = list(province_confirmed_case)
```

```python
for i in nan_indices:
  unique_province.pop(i)
  province_confirmed_case.pop(i)

unique_province
```

```python
plt.figure(figsize=(30,30))
plt.barh(unique_countries,country_confirmed_case)
plt.show()
```
